```python
# Function to load and plot raw vs clean data
def plot_raw_vs_clean(subject, records):
    raws = []
    for record in records:
        edf_path = os.path.join(base_path, subject, f'{subject}{record}.edf')
        raw = mne.io.read_raw_edf(edf_path, preload=True)
        raws.append(raw)
    raw = mne.concatenate_raws(raws)
    
    # Set average reference
    raw.set_eeg_reference('average', projection=True)
    
    # Plot raw data
    fig_raw = raw.plot(n_channels=5, scalings='auto', title=f'Raw EEG Signal - {subject}{record}', show=False)
    
    # Apply a notch filter to remove power line noise
    raw.notch_filter(freqs=50, fir_design='firwin')
    
    # Apply a band-pass filter to clean the data
    raw.filter(4., 30., fir_design='firwin')
    
    # Plot clean data
    fig_clean = raw.plot(n_channels=5, scalings='auto', title=f'Clean EEG Signal - {subject}{record}', show=False)
    plt.show()
    plt.close('all')
    return
```

```python
for i in range(1, 2):
    plot_raw_vs_clean(subjects[i], records)
```

```python
import seaborn as sns

# Function to plot channel-wise data with 6 channels each in separate plots using seaborn
def plot_channel_wise(subject, records, color='b'):
    raws = []
    for record in records:
        edf_path = os.path.join(base_path, subject, f'{subject}{record}.edf')
        raw = mne.io.read_raw_edf(edf_path, preload=True)
        raws.append(raw)
    raw = mne.concatenate_raws(raws)
    
    raw.set_eeg_reference('average', projection=True)
    # Apply a band-pass filter to clean the data
    raw.filter(4., 30., fir_design='firwin')
    
    # Select the first 6 channels
    fig, axes = plt.subplots(6, 1, figsize=(15, 12))
    for i, ax in enumerate(axes):
        data, times = raw[i+3, :]
        sns.lineplot(x=times, y=data[0], ax=ax, color=color)
        ax.set_title(f'Channel {raw.ch_names[i]}')
    plt.tight_layout()
    plt.show()
    return fig
```

```python
color_list = ['r']
for i, color in enumerate(color_list):
    fig = plot_channel_wise(subjects[i], records, color)
    # plt.show()
    plt.close('all')
```
